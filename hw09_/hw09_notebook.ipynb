## Помимо этого задания был ещё поставлен вопрос: "Почему, если разобраться, у критерия Джини целых две формулы?"

### Ответ: 

Если я правильно понял, мы говорим о формулах:
- Gini = 1−∑Pi^2
- Gini = ∑Pi(1−Pi)

Но по сути, ввиду того, что ∑Pi=1, формулы эквивалентны. Связаны различия с тем, что в 1912 году верхнюю формулу вывел Коррадо Джини для измерения неравенства дохов.
А в 1984 году Лео Брейнман, не зная об этой формуле, пытался придумать интуитивно понятную меру нечистоты, которую можно объяснить студентам и инженерам. Он писал: "Если я случайно выберу объект из узла и случайно припишу ему класс (с вероятностью, равной доле классов в узле), то с какой вероятностью я ошибусь?". Оказалось, что с вероятностью, записанной в формуле ниже.
