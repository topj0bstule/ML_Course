```python
y = df.Price
X = df.drop('Price', axis=1)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=0.2
)
```

Теперь применим encoding:

```python
categorical_cols = X.select_dtypes(include='object').columns.to_list()

target_cols = ['Model']
onehot_cols = ['Color', 'Fuel_Type']

preprocessor = ColumnTransformer([
    ('target', TargetEncoder(smoothing=10), target_cols),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False), onehot_cols)
], remainder='passthrough')
```

### 2. Обучение моделей

Продолжаем обучение тут: строим Pipeline, вызываем GridSearchCV, отдельно учим модель с одним деревом, после - отвечаем на вопросы.

```python
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('rf', RandomForestRegressor(random_state=42))
])

m = X.shape[1]

param_grid = {
    'rf__n_estimators': [100, 200, 400],
    'rf__max_depth': [10, 30, 50, 100],
    'rf__min_samples_split': [2, 5],
    'rf__min_samples_leaf': [1, 4, 10, 20, 50],
    'rf__max_features': ['sqrt', 'log2', int(m/3)]
} # хотел бы больше, но мой ноут плачет от боли

grid_search = RandomizedSearchCV(
    pipeline,
    param_grid,
    n_iter=50,
    cv=5, n_jobs=-1, scoring='r2'
)

grid_search.fit(X_train, y_train)
```

```python
print(f"Лучшие параметры: {grid_search.best_params_}")
print(f"Лучший R2 на CV: {grid_search.best_score_}")

rf_model = grid_search.best_estimator_
```

Победила модель с 400 деревьями, макс. глубиной - 30 (соответствует теории: большие деревья).

Интересно, что число признаков - sqrt(m) справилось лучше, чем m/3 (не соотв. теории).

Теперь обучим одно дерево с макс. глубиной 100 и посмотрим, как справится оно:

```python
pipeline_tree = Pipeline([
    ('preprocessor', preprocessor),
    ('tree', DecisionTreeRegressor(random_state=42))
])

param_grid_tree = {
    'tree__max_depth': [50],
    'tree__min_samples_split': [2, 5, 10, 20],
    'tree__min_samples_leaf': [1, 4, 10, 20],
}

grid_search_tree = GridSearchCV(
    pipeline_tree,
    param_grid_tree, 
    cv=5, n_jobs=-1, scoring='r2'
)

grid_search_tree.fit(X_train, y_train)
```

```python
print(f"Лучшие параметры дерева: {grid_search_tree.best_params_}")
print(f"Лучший R² на CV: {grid_search_tree.best_score_:.4f}")
```

Теперь, оценив качество обучения модели на Train, для дальнейшего сравнения с linear regression посмотрим результат r2 и других метрик на test выборке.

```python
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
y_pred = rf_model.predict(X_test)

print("=== Метрики модели случайный лес===")
print(f"Test  R²: {r2_score(y_test, y_pred):.4f}")
print(f"Test  MAE: {mean_absolute_error(y_test, y_pred):.0f}")
print(f"Test  RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.0f}")
print(f"Test  MAPE: {mape(y_test, y_pred):.2f}%")
print(f"Test MSE: {mean_squared_error(y_test, y_pred)}")
```
